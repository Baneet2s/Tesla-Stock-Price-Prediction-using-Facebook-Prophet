DATA PREPARATION

```python
df
```

```python
col = ['Date','Close']
newDF = pd.DataFrame(df, columns=col)
newDF
```

```python
prophet_df = newDF.rename(columns={'Date':'ds', 'Close': 'y'})
prophet_df
```

FACEBOOK PROPHET MODEL CREATION

```python
m = Prophet()
m.fit(prophet_df)
```

FORECASTING

```python
future=m.make_future_dataframe(periods=30)
forecast=m.predict(future)
forecast
```

```python
px.line(forecast, x = 'ds', y = 'yhat')
```

```python
figure1 = m.plot(forecast, xlabel = 'ds', ylabel = 'y')
```

```python
figure2 = m.plot_components(forecast)
```

```python
# Plot actual vs predicted values
prophet_df["ds"] = pd.to_datetime(prophet_df["ds"])
forecast["ds"] = pd.to_datetime(forecast["ds"])

plt.figure(figsize=(10, 6))
plt.plot(prophet_df["ds"], prophet_df["y"], label="Actual Values", color='blue')
plt.plot(forecast["ds"], forecast["yhat"], color="black", label="Predicted Values")
plt.fill_between(forecast["ds"], forecast["yhat_lower"], forecast["yhat_upper"], 
                 color='orange', alpha=0.3, label="Confidence Interval")

plt.title("Actual vs Predicted Values")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
actual = df["y"].values
predicted = forecast["yhat"][:len(df)].values
mae = mean_absolute_error(actual, predicted)
mse = mean_squared_error(actual, predicted)
rmse = np.sqrt(mse)

mare = np.mean(np.abs((actual - predicted) / (actual + 1e-8)))
smape = np.mean(2 * np.abs(predicted - actual) / (np.abs(actual) + np.abs(predicted))) * 100
r2 = r2_score(actual, predicted)
accuracy = 100 - mae
acc1 = 100 - mare

# Display evaluation metrics
print("\n___________________________________________")
print("Prediction Accuracy and Performance Report")
print("___________________________________________")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Relative Error (MARE): {mare:.4f}")
print(f"Symmetric Mean Absolute Percentage Error (SMAPE): {smape:.2f}%")
print(f"Accuracy (based on MAE): {accuracy:.2f}%")
print(f"Accuracy (based on MARE): {acc1:.2f}%")

```
